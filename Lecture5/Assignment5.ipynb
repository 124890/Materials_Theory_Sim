```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.colors as mcolors
import matplotlib.cm as cm
import matplotlib.colorbar as colorbar
import matplotlib.ticker as ticker
import matplotlib.patheffects as path_effects
import matplotlib.lines as mlines
#import mpatches
import matplotlib as mpl
import matplotlib.gridspec as gridspec
import matplotlib.colors as mcolors
import matplotlib.cm as cm
import matplotlib.colorbar as colorbar
import matplotlib.ticker as ticker
import matplotlib.patheffects as path_effects
import matplotlib.lines as mlines
import matplotlib.patches as patches

```

```python
data1 = np.loadtxt('1 - 700 heat loss plot data.txt')
data2 = np.loadtxt('1 - 700 skin depth plot data.txt')
```

```python
# Plot as 1x2 grid with skin depth and heat loss plots side by side
fig, ax = plt.subplots(1, 2, figsize=(20, 5))
plt.subplots_adjust(wspace=0.1)

# Plot 1: Skin depth (1 - 700)
ax[0].plot(data2[:, 0], data2[:, 1], color='red', label='700 nm')
ax[0].set_xlabel('Wavelength (m)')
ax[0].set_ylabel('Skin depth (nm)')

# Plot 2: Heat loss (1 - 700)
ax[1].plot(data1[:, 0], data1[:, 1], color='blue', label='700 nm')
ax[1].set_xlabel('Wavelength (m)')
ax[1].set_ylabel('Heat loss (W)')

# Add red A) and B) labels
ax[0].text(-0.09, 1, 'A)', transform=ax[0].transAxes, fontsize=16, fontweight='bold', va='top', color='red')
ax[1].text(-0.09, 1, 'B)', transform=ax[1].transAxes, fontsize=16, fontweight='bold', va='top', color='red')

#add minor tick marks and make them longer
ax[0].xaxis.set_minor_locator(ticker.AutoMinorLocator())
ax[0].yaxis.set_minor_locator(ticker.AutoMinorLocator())
ax[1].xaxis.set_minor_locator(ticker.AutoMinorLocator())
ax[1].yaxis.set_minor_locator(ticker.AutoMinorLocator())

ax[0].tick_params(which='minor', length=5)
ax[1].tick_params(which='minor', length=5)

#make major tick marks longer
ax[0].tick_params(which='major', length=10)
ax[1].tick_params(which='major', length=10)

#make all ticks thicker
ax[0].tick_params(which = 'both', width=1.5)
ax[1].tick_params(which = 'both', width=1.5)
```

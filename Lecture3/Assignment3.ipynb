```python
import numpy as np

Potential = [0, 4, 8, 12, 16, 20]
Displacement = [0, 16.579, 33.157, 49.736, 66.315, 82.894]

# Plot the data
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 4))

plt.plot(Potential, Displacement, 'x', color='red')
plt.xlabel('Potential (V)', fontsize=14)
plt.ylabel('Displacement (nm)',fontsize = 14)
plt.title('Displacement vs Potential')
plt.xlim(0, 22)

# Set major ticks every 2, minor ticks every 1
plt.xticks(range(0, 21, 2))


plt.grid(False)
plt.minorticks_on()

# Add a trendline
plt.ylim(0, 85)
m, b = np.polyfit(Potential, Displacement, 1)
plt.plot(Potential, [m * i + b for i in Potential])

# Add the trendline to the legend
plt.legend(['Displacement', 'Trendline: y = {0:.2f}x {1:.2f}'.format(m, b)], fontsize=12)
plt.tick_params(which='minor', direction='out')
plt.tick_params(which='major', direction='out')

#make ticks longer
plt.tick_params(which='major', length=8)
plt.tick_params(which='minor', length=5)


plt.show()
```
